```python
from random import random

values = [random() for i in range(20)]
x = random()

values.sort()

matching_indices = [index for index, value in enumerate(values) if value >= x]

print("Sorted values:", values)
print("x:", x)

if matching_indices:
    print("First matching index:", matching_indices[0])
else:
    print("No matching index found (all values < x).")
```
